```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
values_dict = {
    '100':{
        'mean':0,
        'std_deviation':0
    },
    '1000':{
        'mean':0,
        'std_deviation':0
    },
    '5000':{
        'mean':0,
        'std_deviation':0
    }
}
```

```python
# Read the SYN scan packets
ports_100_df=pd.read_csv('../CsvFiles/filtered_nmap -sS -v -F.csv')
ports_1000_df=pd.read_csv('../CsvFiles/filtered_nmap -sS -v.csv')
ports_5000_df=pd.read_csv('../CsvFiles/filtered_nmap -sS -v --top-ports 5000.csv')
```

```python
# Read the TCP connect scan packets
st_ports_100_df = pd.read_csv('../CsvFiles/filtered_nmap -sT -v -F.csv')
st_ports_1000_df = pd.read_csv('../CsvFiles/filtered_nmap -sT -v.csv')
st_ports_5000_df = pd.read_csv('../CsvFiles/filtered_nmap -sT -v --top-ports 5000.csv')
```

```python
# Clean data
# -sS
ss_clean_ports_100_df = ports_100_df.drop_duplicates(subset=['Destination Port'])
ss_clean_ports_1000_df = ports_1000_df.drop_duplicates(subset=['Destination Port'])
ss_clean_ports_5000_df = ports_5000_df.drop_duplicates(subset=['Destination Port'])

# -sT
st_clean_ports_100_df = st_ports_100_df.drop_duplicates(subset=['Destination Port'])
st_clean_ports_1000_df = st_ports_1000_df.drop_duplicates(subset=['Destination Port'])
st_clean_ports_5000_df = st_ports_5000_df.drop_duplicates(subset=['Destination Port'])
```

```python
# Destination ports
# -sS
ss_dp_ports_100_df= ss_clean_ports_100_df['Destination Port']
ss_dp_ports_1000_df= ss_clean_ports_1000_df['Destination Port']
ss_dp_ports_5000_df= ss_clean_ports_5000_df['Destination Port']

# -sT
st_dp_ports_100_df= st_clean_ports_100_df['Destination Port']
st_dp_ports_1000_df= st_clean_ports_1000_df['Destination Port']
st_dp_ports_5000_df= st_clean_ports_5000_df['Destination Port']
```

# SYN scan of 100 ports

```python
# Calculate statistical measures
port_column = ss_dp_ports_100_df

# Mean (average)
mean = port_column.mean()

# Median (middle value)
median = port_column.median()

# Mode (most common value)
mode = port_column.mode().values[0]

# Minimum and Maximum
min_port = port_column.min()
max_port = port_column.max()

# Standard Deviation
std_deviation = port_column.std()

# Variance
variance = port_column.var()

# Summary Statistics
summary_statistics = port_column.describe()
values_dict['100']['mean'] = mean
values_dict['100']['std_deviation'] = std_deviation

# Print or use the calculated values
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Minimum Port:", min_port)
print("Maximum Port:", max_port)
print("Standard Deviation:", std_deviation)
print("Variance:", variance)
print("Summary Statistics:\n", summary_statistics)
```

## Plot for 100 ports

```python
# Port Distribution of 100 port scan
# Given statistics
mean_100 = 5657.1
median_100 = 1069.5
mode_100 = 7
standard_deviation_100 = 11794.444911812414

data = st_dp_ports_100_df
plt.hist(data, bins=50, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_100, standard_deviation_100)
plt.plot(x, p, 'k', linewidth=2)

# Add vertical lines for mean, median, and mode
plt.axvline(mean_100, color='b', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_100, color='r', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(mode_100, color='m', linestyle='dashed', linewidth=2, label='Mode')

# Set labels and title
plt.xlabel('Ports')
plt.ylabel('Probability Density')
# plt.title('Distribution Plot')

# Add legend
plt.legend()

plt.savefig('../../images/PDF of top 100 ports with statistics Measures.png', dpi=600)
# Show the plot
plt.show()
```

# Summary

```python
def plot_distribution(ax, mean, median, mode, std_dev, data, title):
    ax.hist(data, bins=50, density=True, alpha=0.6, color='g', label='Histogram')

    # Plot the normal distribution curve
    xmin, xmax = ax.get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mean, std_dev)
    ax.plot(x, p, 'k', linewidth=2)

    # Add vertical lines for mean, median, and mode
    ax.axvline(mean, color='b', linestyle='dashed', linewidth=2, label='Mean')
    ax.axvline(median, color='r', linestyle='dashed', linewidth=2, label='Median')
    ax.axvline(mode, color='m', linestyle='dashed', linewidth=2, label='Mode')

    # Set labels and title
    ax.set_xlabel('Ports')
    ax.set_ylabel('Probability Density')
    ax.set_title(title)

    # Add legend
    ax.legend()

# Given statistics for each port scan
mean_100 = 5657.1
median_100 = 1069.5
mode_100 = 7
standard_deviation_100 = 11794.444911812414

mean_1000 = 8634.848
median_1000 = 3720.0
mode_1000 = 1
standard_deviation_1000 = 13448.483415354654

mean_5000 = 12118.3094
median_5000 = 3798.5
mode_5000 = 1
standard_deviation_5000 = 17368.5302826954

# Given data for each port scan
data_100 = st_dp_ports_100_df
data_1000 = st_dp_ports_1000_df
data_5000 = st_dp_ports_5000_df

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot distributions for each port scan
plot_distribution(axs[0], mean_100, median_100, mode_100, standard_deviation_100, data_100, 'Port Distribution of 100 port scan')
plot_distribution(axs[1], mean_1000, median_1000, mode_1000, standard_deviation_1000, data_1000, 'Port Distribution of 1000 port scan')
plot_distribution(axs[2], mean_5000, median_5000, mode_5000, standard_deviation_5000, data_5000, 'Port Distribution of 5000 port scan')

# Adjust layout
plt.tight_layout()

plt.savefig('../../images/ports distribution with statistics measures.png', dpi=600)
# Show the plot
plt.show()
```
